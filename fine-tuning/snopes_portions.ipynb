```python
import pandas as pd
import re
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
snps_df = pd.read_csv('/content/drive/MyDrive/Plain_DeClarE/snopes.tsv', sep='\t', header=None)
```

```python
snps_df.columns = ['label', 'claim_id', 'claim', 'text', 'article_source']  # label: credibility, text: article body
```

```python
categories = snps_df['claim_id'].str.lower().apply(lambda x: re.split('_|-', x)).explode().drop_duplicates().tolist()
```

```python
len(categories)
```

```python
category_count = {}
for item in categories:
  category_count[item] = 0
for row in range(len(snps_df)):
  row_cats = re.split('_|-', snps_df['claim_id'].iloc[row].lower())
  for item in row_cats:
     category_count[item] += 1
```

```python
category_count
```

```python
sorted_category_count = dict(sorted(category_count.items(), key=lambda x: x[1], reverse=True))
```

```python
sorted_category_count
```

```python
snps_df['politics_category'] = snps_df['claim_id'].str.lower().apply(lambda x: re.split('_|-', x)).apply(lambda x: 'politics' in x)
```

```python
snps_df.head(10)
```

```python
snps_nonpltcs = snps_df[snps_df.politics_category==False]
```

```python
len(snps_nonpltcs)
```

```python
snps_nonpltcs.drop(['politics_category'], axis=1, inplace=True)
```

```python
snps_nonpltcs['label'].value_counts()
```

```python
snps_nonpltcs['label'].replace({'mostly true':'true', 'mostly false':'false'}, inplace=True)
```

```python
snps_nonpltcs.head()
```
